```python
import pandas as pd
import matplotlib.pyplot as plt
import os

def graficar_precipitacion_anual(csv_path):


    df = pd.read_csv(csv_path, parse_dates=['valid_time'])


    df['tp'] = df['tp'] * 1000  

    df['año'] = df['valid_time'].dt.year

    media_anual = df.groupby('año')['tp'].mean()

    resumen_csv = "media_anual_precipitacion.csv"
    media_anual.to_csv(resumen_csv, header=["precipitacion_media_mm"])
    print(f"Archivo guardado con medias anuales: {resumen_csv}")

    plt.figure(figsize=(10, 5))
    media_anual.plot(marker='o')
    plt.title("Evolución de la media anual de precipitación (mm)")
    plt.xlabel("Año")
    plt.ylabel("Precipitación media (mm)")
    plt.grid(True)
    plt.tight_layout()

    output_img = "media_anual_precipitacion.png"
    plt.savefig(output_img)
    plt.show()
    print(f"Gráfico guardado como: {output_img}")
```

```python
graficar_precipitacion_anual("precipitacion_mediterraneo.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

def analizar_precipitacion(csv_path):
    

    df = pd.read_csv(csv_path, parse_dates=['valid_time'])


    df['tp'] = df['tp'] * 1000
    df['año'] = df['valid_time'].dt.year

    media_anual = df.groupby('año')['tp'].mean()
    total_anual = df.groupby('año')['tp'].sum()

    media_anual.to_csv("media_anual_precipitacion.csv", header=["precipitacion_media_mm"])
    total_anual.to_csv("total_anual_precipitacion.csv", header=["precipitacion_total_mm"])
    print("CSVs generados:")
    print("- media_anual_precipitacion.csv")
    print("- total_anual_precipitacion.csv")

    plt.figure(figsize=(10, 5))
    media_anual.plot(marker='o', color='royalblue')
    plt.title("Media anual de precipitación (mm)")
    plt.xlabel("Año")
    plt.ylabel("Precipitación media por celda (mm)")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig("grafico_media_anual.png")
    plt.show()

    plt.figure(figsize=(10, 5))
    total_anual.plot(marker='o', color='seagreen')
    plt.title("Precipitación total anual acumulada en la región (mm)")
    plt.xlabel("Año")
    plt.ylabel("Precipitación total (mm)")
    plt.grid(True)
    plt.tight_layout()
    plt.savefig("grafico_total_anual.png")
    plt.show()

    print("Gráficos guardados:")
    print("- grafico_media_anual.png")
    print("- grafico_total_anual.png")
```

```python
analizar_precipitacion("precipitacion_mediterraneo.csv")
```
